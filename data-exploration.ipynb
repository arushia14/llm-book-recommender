```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# dealing with missing values here

ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("columns")
plt.ylabel("missing")

plt.show()
```

```python
import numpy as np

books["missing_description"]= np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
# checking to see if missing descriptions have any correlation with other fields that will skew it a direction

columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize = (8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})
plt.show()
```

```python
# since there is no correlation between missing description with other values, removing the missing vals since they make less
# than 5% of our dataset

book_missing = books.dropna(subset=["description", "num_pages", "average_rating", "published_year"])

book_missing
```

```python
# categories seem to be very specific so checking what is going on there

book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
# will deal with skewed categories with LLM
# here, trying to remove any unhelpful descriptions with few words or so that won't be useful
book_missing = books[~books["description"].isna() &
                     ~books["num_pages"].isna() &
                     ~books["average_rating"].isna() &
                     ~books["published_year"].isna()].copy()

book_missing["words_in_description"] = book_missing["description"].str.split().str.len()

book_missing
```

```python
# checking what very short (up to 4 words of description) looks like
book_missing.loc[book_missing["words_in_description"].between(1,4), "description"]
```

```python
# that was too short so let's see longer descriptions
book_missing.loc[book_missing["words_in_description"].between(25, 34), "description"]
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
book_missing_25_words
```

```python
# since many of these books don't have subtitle, we can create a new field title_and_subtitle where it is title: subtitle

book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(),
             book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
book_missing_25_words
```

```python
# tagging each description with a unique identifier for LLM

book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
book_missing_25_words
```

```python
# finally dropping all the columns we don't need and have the cleaned data

(
    book_missing_25_words.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"],axis=1)
    .to_csv("books_cleaned.csv", index=False)
 )
```
